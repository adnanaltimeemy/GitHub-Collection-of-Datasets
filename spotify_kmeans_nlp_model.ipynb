```python
# =========================
# 1. Imports
# =========================
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# =========================
# 2. Download NLTK resources (run once)
# =========================
nltk.download('punkt')
nltk.download('stopwords')

# =========================
# 3. Example: Load your data
# =========================
# Replace this with your actual file
# df = pd.read_csv('your_file.csv')

# Example dummy data (remove if you already have df)
df = pd.DataFrame({
    'text': [
        "Hello World! This is a TEST 123.",
        "Cleaning text, removing stopwords and numbers!",
        "NLTK is useful for NLP tasks."
    ]
})

# =========================
# 4. Clean column names (IMPORTANT)
# =========================
df.columns = df.columns.str.strip().str.lower()

# =========================
# 5. Stopwords
# =========================
stop_words = set(stopwords.words('english'))

# =========================
# 6. Cleaning function
# =========================
def clean_text(text):
    # Handle missing values
    if pd.isna(text):
        return ""
    
    # lowercase
    text = text.lower()
    
    # remove numbers
    text = re.sub(r'\d+', '', text)
    
    # remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # tokenize
    tokens = word_tokenize(text)
    
    # remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    return " ".join(tokens)

# =========================
# 7. Apply cleaning
# =========================
df['clean_text'] = df['text'].apply(clean_text)

# =========================
# 8. View results
# =========================
print(df[['text', 'clean_text']].head())
```

```python
vectorizer = TfidfVectorizer(max_features=2000)

X = vectorizer.fit_transform(df['clean_text'])

print(X.shape)
```
